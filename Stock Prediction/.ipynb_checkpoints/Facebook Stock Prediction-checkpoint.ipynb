```python
#import panda packages
import pandas as pd

#Read Data from CSV file And Store It IN DATAFRAME
df = pd.read_csv('FB.csv')

#print TOP 10 RECORD
df.head(10)
```

```python
#setting index as date
df['Date'] = pd.to_datetime(df.Date,format='%Y-%m-%d')
df.index = df['Date']

#to plot the graph
import matplotlib.pyplot as plt
%matplotlib inline

# Plot everything by leveraging the very powerful matplotlib package
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(df['Close'], label='Share Closing Price')

ax.set_xlabel('DATES - YEAR/MONTH/DAY',fontsize=14, color='r')
ax.set_ylabel('SHARE CLOSING PRICE',fontsize=14, color='r')
ax.set_title('ORIGINAL CLOSING PRICE PLOT', fontsize=20, color='g')
ax.legend()
```

```python
#creating dataframe with date and the target variable
data = df.sort_index(ascending=True, axis=0)
train_data = pd.DataFrame(index=range(0,len(df)),columns=['Date', 'Close'])

for i in range(0,len(data)):
     train_data['Date'][i] = data['Date'][i]
     train_data['Close'][i] = data['Close'][i]

train_data.to_csv('FB_MovingAvgData.csv')
train_data.head(10)
```

```python
#Last 3 months Prediction
#splitting into train and test
df_test = pd.read_csv('FB_Testdata.csv')
#setting index as date
df_test['Date'] = pd.to_datetime(df_test.Date,format='%Y-%m-%d')
df_test.index = df_test['Date']
#creating dataframe with date and the target variable
data = df_test.sort_index(ascending=True, axis=0)
test_data = pd.DataFrame(index=range(0,len(df_test)),columns=['Date', 'Close'])

for i in range(0,len(data)):
     test_data['Date'][i] = data['Date'][i]
     test_data['Close'][i] = data['Close'][i]

test_data.to_csv('FB_test_MovingAvgData.csv')
```

```python
# Shape Return No of ROWS AND COLUMN OF A DATAFRAME
train_data.shape, test_data.shape
```

```python
# TO Check the First and Last Record Value
train_data['Date'].min(), train_data['Date'].max(), test_data['Date'].min(), test_data['Date'].max()
```

```python
#make predictions
stockclose_preds = []

# range value taken from test.shape
test_shape_no_of_records = 32
#Taking No of Record to Calculate Avg
no_of_record = 40

for i in range(0,test_shape_no_of_records):
    if i < no_of_record:
        a = train_data['Close'][len(train_data)-no_of_record+i:].sum() + sum(stockclose_preds)
    else:
        c = stockclose_preds[-no_of_record:]
        a = sum(c)
    
    b = a/no_of_record
    stockclose_preds.append(b)    
```

```python
date_list = []

date_list = list(test_data.Date) # TAKE DATE FROM TEST DATA FRAME AND CONVERT IT TO LIST
#print (stcokclose_preds)
#print (date_list)
 
dict = {'Date':date_list,'Predict_Close': stockclose_preds} # CREATE DICTIONARY USING LIST
#print(dict)

predict_df = pd.DataFrame(dict)  #CONVERT DICTIONARY TO DATAFRAME
predict_df.to_csv('FB_MovingAvgPredictData.csv')
```

```python
#calculate rmse
#MSE (Mean Squared Error) = np.mean((y_test - y_predtest)**2)
#RMSE (Root Mean Squared Error) = np.sqrt(MSE)

import numpy as np

x = np.power((np.array(test_data['Close'])-stockclose_preds),2)
mse = np.mean(x)
rmse = np.sqrt(mse)
rmse
```

```python
# Plot everything by leveraging the very powerful matplotlib package
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(train_data['Date'],train_data['Close'],label = 'Training Closing Price')
ax.plot(test_data['Date'],test_data['Close'],label = 'Original Closing Price')
ax.plot(predict_df['Date'],predict_df['Predict_Close'],label = 'Predict Closing Price')

ax.set_xlabel('DATES - YEAR/MONTH/DAY',fontsize=14, color='r')
ax.set_ylabel('SHARE CLOSING PRICE',fontsize=14, color='r')
ax.set_title('Training Closing Price VS Original Closing Price VS Predict Closing Price', fontsize=20, color='g')
ax.legend()
```

```python
#Defining MAPE function
def MAPE(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_Predicted))*100
    return mape

print(MAPE(test_data['Close'],stockclose_preds))
```
